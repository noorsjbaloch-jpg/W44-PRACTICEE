## Exercise 1 — Load the data
Load `data/sports_scores.csv` using `np.genfromtxt` with `delimiter=","` and `names=True`.
Then print the column names and how many matches there are.

```python
# YOUR CODE HERE
#data = np.genfromtxt("sports_scores.csv",delimiter=",",names=True)
import os

data = np.genfromtxt("sports_scores.csv", delimiter=",", names=True)

print(data[:5])
print ("column names:",data.dtype.names)
match = data["match"]
points_scored = data["points_scored"]
points_allowed = data["points_allowed"]
rebounds = data["rebounds"]
assists = data["assists"]
```

## Exercise 2 — Points summary
For the `points_scored` column, print the average, highest, and lowest points.

```python
# YOUR CODE HERE

print("Average points;",points_scored.mean())
print ("Highest score:",points_scored.max())
print("Lowest score:",points_scored.min())
```

## Exercise 3 — Winning games
A game is a WIN when `points_scored` is greater than `points_allowed`.
1. Build a boolean mask for wins.
2. Count how many games were won.
3. Print the match numbers that were wins.

```python
# YOUR CODE HERE
import numpy as np
print (np.where(points_scored>points_allowed,"WIN","LOSS"))
count= (points_scored>points_allowed).sum()
print("No. of games won :",int(count))
wins= points_scored >points_allowed
print(match[wins])
```

## Exercise 4 — Best game
Use `argmax` to find the position of the highest-scoring game, then print which match number it was and how many points were scored.

```python
# YOUR CODE HERE
highscore = points_scored.argmax()
print(highscore)
highscore_match = data['match'][highscore]
print("The highest scoring game was" ,int(highscore_match),"with", int(highscore))
```

## Exercise 5 — Label the games
Use `np.where` to label each game `"HIGH"` if points_scored ≥ 105, else `"LOW"`. Print the labels.

```python
# YOUR CODE HERE
points = np.where(points_scored>=105,"HIGH","LOW")
print(points)
```
